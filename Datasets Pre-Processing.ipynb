```python
import pandas as pd
pd.options.display.max_rows = 5000#Changes the number of rows diplayed (default is 60)
pd.set_option('display.precision', 2)
```

```python
makeup_trustworthiness_attractiveness = pd.read_excel('BeautyPremium.xlsx').drop(columns=['makeupdummy','mk','Averagetrust','gendermakeup','averageattract','GroupAttrac','IncreaseAttract','transfer'])

makeup_trustworthiness_attractiveness=makeup_trustworthiness_attractiveness.rename(columns={"player2": "Model", "treatment":'Makeup','male':'Gender','attractiveness':'Attractiveness'})


makeup_trustworthiness_attractiveness = makeup_trustworthiness_attractiveness.replace({'Gender':0,'Makeup':'nomakeup'}, 
                 {'Gender': 'Male','Makeup':'False'},regex=True)
makeup_trustworthiness_attractiveness = makeup_trustworthiness_attractiveness.replace({'Gender':1,'Makeup':'makeup'}, 
                 {'Gender':'Female','Makeup':'True'},regex=True)
#print(makeup_trustworthiness_attractiveness)

makeup_trustworthiness_attractiveness.to_excel('Makeup Trustworthiness and Attractiveness Effects Study.xlsx')

```
